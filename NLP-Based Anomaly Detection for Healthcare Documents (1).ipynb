# 1: OCR - Text Pre-Processing
---

### Extracting Structured Data from OCR-Processed Text

This function, `extract_ocr_data(text)`, uses rule-based regular expressions to extract key fields from unstructured OCR-scanned healthcare claims.

OCR output is often messy, inconsistent, and noisy — so a lightweight, explainable method like regex is a practical first step before applying any machine learning.

#### Fields Extracted:
- **Patient Name** – e.g., `Patient: John Doe`
- **ICD-10 Code** – e.g., `ICD-10: E11.9`
- **MDC Code** – e.g., `MDC: 08` (Major Diagnostic Category)
- **Charge Amount** – e.g., `Charge: $1540.00`
- **Procedure Description** – e.g., `Procedure: Appendectomy`

Each field is extracted using a simple pattern, and if not found, is returned as `None`. The charge amount is also safely converted to a `float` for downstream analysis (e.g., anomaly detection).

#### Why this matters:
By turning messy OCR text into a clean structured format, we can:
- Detect anomalies (e.g., outlier charges or invalid codes)
- Validate data quality
- Feed structured records into machine learning models for fraud detection or claims automation

This is a practical example of combining regular expressions, healthcare domain knowledge, and data engineering to prepare EDI data for further analysis.


````python
def extract_ocr_data(text):
    #rule based extraction
    patient = re.search(r'Patient: (.*?)(?:\n|$)', text, re.IGNORECASE)
    icd10 = re.search(r'ICD-10: (.*?)(?:\n|$)', text, re.IGNORECASE)
    mdc = re.search(r'MDC: (\d+)(?:\n|$)', text, re.IGNORECASE)
    charge = re.search(r'Charge: \$(\d+\.?:\n|$)', text, re.IGNORECASE)
    procedure = re.search(r'Procedure: (.*?)(?:\n|$)', text, re.IGNORECASE)

    return {
    'patient': patient.group(1).strip() if patient else None,
    'icd10': icd10.group(1).strip() if icd10 else None,
    'mdc': mdc.group(1).strip() if mdc else None,
    'charge': float(charge.group(1)) if charge else None,
    'procedure': procedure.group(1).strip() if procedure else None
    }
````

### Applying OCR Extraction and Structuring the Output

# 2: EDI - Text Pre-processing
---
### Parsing EDI 837 Healthcare Claim Messages

This function, `parse_edi_message(message)`, is designed to extract key fields from a simplified **HIPAA 837** EDI healthcare claim.

EDI files are structured but not human-readable, making it important to extract relevant claim data programmatically.

#### Fields Parsed:
- **Claim ID** from the `CLM` segment  
- **Charge Amount** (converted to `float`) from the same `CLM` segment  
- **ICD-10 Diagnosis Code** from the `HI` segment

#### Example Segment Patterns:
- `CLM*12345*1250.00***`  
- `HI*ABK:E11.9`

#### Output:
The function returns a dictionary with:
```python
{
    'claim_id': '12345',
    'charge': 1250.00,
    'icd10': 'E11.9'
}

````python
def parse_edi_message(message):
# Parse HIPAA 837 (simplified)
    clm = re.search(r'CLM\*(\d+)\*(\d+\.?\d*)\*\*\*', message)
    hi = re.search(r'HI\*ABK:(.*?)(?:\n|$)', message)
    return {
        'claim_id': clm.group(1).strip() if clm else None,
        'charge': float(clm.group(2)) if clm else None,
        'icd10': hi.group(1).strip() if hi else None
    }
````

````python
edi_extracted = df['edi_message'].apply(parse_edi_message)
edi_df = pd.json_normalize(edi_extracted)
print(edi_extracted)
print(edi_df)
````

````python
edi_df.columns = [f'edi_{col}' for col in edi_df.columns]
df = pd.concat([df, edi_df], axis=1)
df.head(1)
````

# 3: MDC Validation mapping
---

### Step 3: MDC Validation – Cross-Checking OCR with EDI Data

This step validates the **Major Diagnostic Category (MDC)** code extracted via OCR against the expected MDC derived from structured EDI data.

#### Function: `validate_mdc(df, mdc_mapping)`

- **Input:**
  - `df`: DataFrame containing both OCR-extracted fields (`ocr_mdc`) and parsed EDI fields (`edi_icd10`)
  - `mdc_mapping`: A reference dictionary mapping ICD-10 diagnosis codes to expected MDC values

#### Logic:
1. **Map ICD-10 to Expected MDC:**
   - Using the ICD-10 code parsed from the EDI message, the function adds a new column `expected_mdc` by looking up the code in `mdc_mapping`.

2. **Validate OCR Accuracy:**
   - Compares the expected MDC with the OCR-extracted `ocr_mdc` value.
   - Flags a mismatch with a new Boolean column `mdc_mismatch`.

3. **Anomaly Tagging:**
   - The column `is_anomaly_mdc` explicitly flags any record where the OCR value does not match the expected EDI-derived MDC.

#### Why this matters:
Validating MDC consistency is important for:
- Identifying **potential OCR errors** in scanned forms
- Catching **incorrect medical coding**
- Detecting **fraudulent claims** or **data entry mistakes**

This step demonstrates how EDI and OCR data can be cross-validated to ensure higher data integrity in healthcare claim workflows.

````python
# Step 3: MDC Validation
def validate_mdc(df, mdc_mapping):
    # Map ICD-10 to expected MDC
    df['expected_mdc'] = df['edi_icd10'].map(mdc_mapping)
    # Check if OCR MDC matches expected MDC
    df['mdc_mismatch'] = df['ocr_mdc'] != df['expected_mdc']
    df['is_anomaly_mdc'] = df['mdc_mismatch']
    return df

df = validate_mdc(df, mdc_mapping)
df.head(1)
````

# 4: Procedures NLP-Based Validation with TF-IDF & Isolation Forest
---

### Procedure Description Validation using NLP and Anomaly Detection

This step applies Natural Language Processing (NLP) to OCR-extracted **procedure descriptions** to identify outliers using machine learning.

#### Function: `nlp_validation(df)`

##### Key Steps:

1. **Text Preprocessing:**
   - The `ocr_procedure` field is filled with empty strings if null.
   - All procedure descriptions are stored in a single column: `all_procedures`.

2. **TF-IDF Vectorization:**
   - Applies `TfidfVectorizer` to convert text data into numerical features.
   - Captures the top 50 most relevant terms while removing English stop words.
   - Transforms the text into a vectorized matrix `X_procedures`.

3. **Anomaly Detection with Isolation Forest:**
   - Fits an `IsolationForest` model to the TF-IDF features.
   - Flags rare or unusual procedure descriptions as anomalies.
   - Adds a Boolean column `is_anomaly_procedure` to the DataFrame.

#### Why this matters:
- **OCR errors** in free-text fields like procedures are hard to catch with rule-based logic alone.
- Using NLP allows for **unsupervised anomaly detection** based on the content of procedure descriptions.
- Helps flag potential **fraudulent**, **misclassified**, or **unusual** procedures in healthcare claims data.

This method highlights the power of combining machine learning with unstructured text data to improve data quality and integrity in healthcare claims processing.

````python

def nlp_validation(df):
    # Combine procedure descriptions
    df['all_procedures'] = df['ocr_procedure'].fillna('')
    
    # TF-IDF vectorization
    vectorizer = TfidfVectorizer(max_features=50, stop_words='english')
    X_procedures = vectorizer.fit_transform(df['all_procedures']).toarray()
    
    # Detect anomalies in procedures
    iso_forest = IsolationForest(contamination=0.01, random_state=42)
    df['is_anomaly_procedure'] = iso_forest.fit_predict(X_procedures) == -1
    
    return df, X_procedures

df, X_procedures = nlp_validation(df)
````

NLP with Scikit-learn: Use TF-IDF and Isolation Forest Model to process text fields (diagnosis descriptions) and validate MDC codes.

 * __TF-IDF vectorization:__ Used to convert a collection of text into a matrix of TF-IDF features. Gives more weight to words that are frequent in a document but rare across documents. Also removes englisg stop words that are not useful and sends the text to an array to be used by the isolation forest below which will perform anomoly detection (i.e find outliers). 

 * __Isolation Forest:__ An usupervised learning ensemble model Instead of profiling "normal" data, it isolates anomalies by randomly selecting a feature, then selecting a split value between its min and max. Since anomalies are rare and different, they get isolated faster. 

 * contamination=0.01 tells the model that ~1% of the data is expected to be anomalous. random_state=42 ensures reproducibility.

Isolation forset creates a new column is_anomaly_procedure with True for anomalies, False otherwise. Fits the model on X_procedures (a feature matrix of TF-IDF procedure data) and labels points as anomalies (-1) or normal (1).

# 5: Anomaly Detection on Combined Data
---

### Combined Anomaly Detection using Text and Numerical Features

In this step, we perform **unsupervised anomaly detection** by leveraging both unstructured and structured data from healthcare claims.

---

#### Step 1: NLP-Based Text Feature Analysis

We create a new feature `all_text` by combining:
- `ocr_procedure`: OCR-extracted procedure description
- `ocr_provider`: OCR-extracted provider name (if available)

We then apply **TF-IDF Vectorization** to convert the text into a numeric matrix, capturing the 50 most relevant terms.


````python
def nlp_validation(df):
    # Combine procedure and provider for TF-IDF
    df['all_text'] = df['ocr_procedure'].fillna('') + ' ' + df.get('ocr_provider', '')
    vectorizer = TfidfVectorizer(max_features=50, stop_words='english')
    X_text = vectorizer.fit_transform(df['all_text']).toarray()
    iso_forest = IsolationForest(contamination=0.01, random_state=42)
    df['is_anomaly_text'] = iso_forest.fit_predict(X_text) == -1
    return df, X_text

df, X_text = nlp_validation(df)

# Step 6: Anomaly Detection
def apply_anomaly_detection(df, numerical_cols=['ocr_charge', 'edi_charge'], contamination=0.01):
    X_num = df[numerical_cols].fillna(0)
    scaler = StandardScaler()
    X_num_scaled = scaler.fit_transform(X_num)
    X_combined = np.hstack([X_num_scaled, X_procedures])
    iso_forest = IsolationForest(contamination=contamination, random_state=42)
    df['is_anomaly_combined'] = iso_forest.fit_predict(X_combined) == -1
    df['is_anomaly'] = (df['is_anomaly_mdc'] | df['is_anomaly_procedure'] | 
                        df['is_anomaly_combined'])
    return df


df = apply_anomaly_detection(df)
df['is_anomaly'].value_counts()
````
